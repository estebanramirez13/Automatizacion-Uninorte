```python
#fecha en formato dd/mm/yyyy usando datetime
import datetime
from datetime import datetime
# Convertir la columna 'Fecha registrada' a tipo datetime
# Asumiendo que la fecha está en formato '2025-01-21 14:31:23.781000' y convertirla al formato deseado dd/mm/yyyy
# Primero, asegurémonos de que la columna 'Fecha registrada' sea de tipo string 
da['Fecha registrada'] = da['Fecha registrada'].astype(str)
# Luego, convertimos la columna a tipo datetime
# y especificamos el formato de entrada '%Y-%m-%d %H:%M:%S.%f'
# Finalmente, convertimos al formato deseado 'dd/mm/yyyy'
da['Fecha registrada'] = pd.to_datetime(da['Fecha registrada'], format='%Y-%m-%d %H:%M:%S.%f').dt.strftime('%d/%m/%Y')
da
```

```python
df=pd.read_excel("Calidad_IngenieríaCivil_LaboratorioGeotecnia_V20062023_3+de+julio+de+2025_07.59.xlsx")
df.columns = df.iloc[0]
df = df[1:]
df
```
