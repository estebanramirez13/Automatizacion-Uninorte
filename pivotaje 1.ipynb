```python
import pandas as pd
df=pd.read_excel("Resultados encuesta de satisfacción servicios TIC-noviembre2025 (1).xlsx")
```

```python
def funcion_pivotaje(archivo):
    def detectar_puntos(dato):
      j=0
      for d in dato:
        if ':' in d or d[-1]==':':
          j=j+1
      return True if j==len(dato) else False

    def detectar_columnas_caso(df):
        casos=[]
        for columns in df.columns:
          if 'caso' in columns.strip().lower():
            casos.append(columns)
        return casos

    def valores_respuesta_replace(datos):
      datos_unicos=set([str(x) for x in datos])
      dicionario_respuesta={}
      for d in datos_unicos:
        if '5' in d:
          dicionario_respuesta[d]=5
        elif '4' in d:
          dicionario_respuesta[d]=4
        elif '3' in d:
          dicionario_respuesta[d]=3
        elif '2' in d:
          dicionario_respuesta[d]=2
        elif '1' in d:
          dicionario_respuesta[d]=1
      return dicionario_respuesta

    def detectar_columnas_respuesta(df):
      for columns in df.columns:
        if 'respuesta' in columns.strip().lower():
          return columns

    def detectar_column_pregnta(df):
      for columns in df.columns:
        i=0
        if 'pregunta' in columns.strip().lower():
          i=i+1
      if i>1:
          l_pregunta=[x for x in df.columns if 'pregunta' in x.strip().lower()]
          L_valor=[x for x in l_pregunta if detectar_puntos(x)==True]
          column_pregunta=l_pregunta[0]
      else:
        column_pregunta=[x for x in df.columns if 'pregunta' in x.strip().lower()][0]
      return column_pregunta

    def pivotaje_data(indexc,columnsc,valuesc,df):
      data = df.pivot_table(index=indexc, columns=columnsc, values=valuesc, aggfunc='first').reset_index()
      return data
    iindexc=detectar_columnas_caso(archivo)
    icolumnsc=detectar_column_pregnta(archivo)
    ivaluesc=detectar_columnas_respuesta(archivo)
    data=pivotaje_data(iindexc,icolumnsc,ivaluesc,archivo)
    return data
funcion_pivotaje(df)
```

```python
def valores_respuesta_replace(datos):
  datos_unicos=set([str(x) for x in datos])
  dicionario_respuesta={}
  for d in datos_unicos:
    if '5' in d:
       dicionario_respuesta[d]=5
    elif '4' in d:
       dicionario_respuesta[d]=4
    elif '3' in d:
       dicionario_respuesta[d]=3
    elif '2' in d:
       dicionario_respuesta[d]=2
    elif '1' in d:
      dicionario_respuesta[d]=1
  return dicionario_respuesta
```

```python
df=df.astype(str).apply(lambda x: x.str.strip())
data = df.pivot_table(index=['N_Caso', 'Tipo_caso'], columns='Pregunta', values='Respuesta', aggfunc='first').reset_index()
data=data.replace(valores_respuesta_replace(df['Respuesta']))
data=data.astype(str).apply(lambda x: x.str.strip())
data.columns.name = None # Eliminar el index columna pregunta
display(data)
data.to_excel('CSUNOV.xlsx')
```
